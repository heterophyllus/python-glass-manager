```python

from glass import *
import glob
import matplotlib.pyplot as plt
%matplotlib inline

files = glob.glob("./AGF/*.agf")

manager = GlassCatalogManager()
manager.load_agf_files(files)
```

```python
def list_refractive_index():
    catalog = manager.catalogs[1]
    for g in catalog.glasses:
        nd = g.refractive_index(SpectralLine.d/1000.0)
        print(g.product_name, nd)

list_refractive_index()
```

```python

# This function plots standard nd-vd map
def plot_glassmap():
    for catalog in manager.catalogs:
        xlist = []
        ylist = []
        slist = []

        for g in catalog.glasses:
            x = g.abbe_d()
            y = g.refractive_index(SpectralLine.d/1000.0)
            s = g.product_name
            xlist.append(x)
            ylist.append(y)
            slist.append(s)

            # text label for each points
            #plt.text(x,y,s)
            
        plt.scatter(xlist,ylist, label= catalog.supplier)

    plt.legend(loc='upper left', bbox_to_anchor=(1,1))
    plt.gca().invert_xaxis()
    plt.ylim(1.4, 2.1)
    plt.show()

# run the function
plot_glassmap()
```
